```python
#1.

A database is a structured collection of data that is organized, stored, and managed in a way that allows for efficient retrieval, modification, and analysis.

SQL Databases:
SQL databases are based on the relational model, where data is organized into tables with rows and columns, and relationships between tables are established through keys. 

NoSQL Databases:
NoSQL databases depart from the rigid structure of SQL databases and are designed to handle large volumes of unstructured or semi-structured data. 
```

```python
#2. 

DDL stands for "Data Definition Language," and it's a subset of SQL that is used to define and manage the structure of a database, including creating, modifying, and deleting database objects like tables, indexes, and constraints. 

CREATE: The CREATE statement is used to create new database objects, such as tables, indexes, views, or schemas. 
It defines the structure and characteristics of the object being created.

Example:

CREATE TABLE Employees (
    EmployeeID INT PRIMARY KEY,
    FirstName VARCHAR(50),
    LastName VARCHAR(50),
    Department VARCHAR(50)
);

DROP: The DROP statement is used to delete existing database objects, including tables, indexes, views, or schemas. 
This permanently removes the object and its associated data from the database.

Example:
    
DROP TABLE Employees;

ALTER: The ALTER statement is used to modify an existing database object's structure, such as adding or deleting columns, changing data types, or modifying constraints.

Example:

ALTER TABLE Employees
ADD Email VARCHAR(100);

TRUNCATE: The TRUNCATE statement is used to remove all rows from a table, effectively resetting the table's data while keeping its structure intact. 
Unlike the DELETE statement, TRUNCATE is faster because it doesn't log individual row deletions.

Example:
    
TRUNCATE TABLE Employees;
```

```python
#3.

DML stands for "Data Manipulation Language," which is a subset of SQL (Structured Query Language) used to manipulate data stored in a relational database. 

INSERT: The INSERT statement is used to add new rows or records into a database table. It allows you to provide values for each column in the new row. 
Here's an example:
Suppose we have a table called "Students" with columns "StudentID," "FirstName," "LastName," and "Age." 
We want to insert a new student into this table.

INSERT INTO Students (StudentID, FirstName, LastName, Age)
VALUES (101, 'John', 'Doe', 20);

UPDATE: The UPDATE statement is used to modify existing records in a table. 
You can update one or more columns of a specific row based on a certain condition. Here's an example:

Let's say we want to update the age of the student with StudentID 101 to 21.

UPDATE Students
SET Age = 21
WHERE StudentID = 101;

DELETE: The DELETE statement is used to remove one or more rows from a table based on a specified condition. Here's an example:

Suppose we want to delete the record of the student with StudentID 101 from the "Students" table.

DELETE FROM Students
WHERE StudentID = 101;
```

```python
#4.

DQL stands for "Data Query Language," and it refers to the subset of SQL (Structured Query Language) that is used for querying and retrieving data from a relational database. 

One of the fundamental commands in DQL is the SELECT statement, which is used to retrieve data from one or more tables in a database. 
The SELECT statement allows you to specify the columns you want to retrieve, the table you want to retrieve data from, and optional conditions to filter the data.

Suppose we have a simple database with a table called "Employees" that contains information about employees in a company. 
The table might have columns like "EmployeeID," "FirstName," "LastName," "Department," and "Salary."

To retrieve the names and salaries of all employees in the "IT" department, you could use the following SQL SELECT statement:
    
SELECT FirstName, LastName, Salary
FROM Employees
WHERE Department = 'IT';
```

```python
#5.

Primary Key:
A primary key serves as a distinctive identifier for individual rows within a table in a relational database. 
It guarantees that each row possesses a unique identification, preventing duplicates and maintaining data accuracy. 
A primary key is also used as a reference point for establishing relationships with other tables. 

Foreign Key:
A foreign key establishes a connection between two tables in a relational database. 
It involves a column or columns in one table that refer to the primary key in another table. 
This linkage forms relationships and dependencies between tables, aiding in the organization and retrieval of data across related entities. 
```

```python
#6. 

import mysql.connector


db_config = {
    "host": "localhost",
    "user": "your_username",
    "password": "your_password",
    "database": "your_database_name"
}

# Connect to the database
connection = mysql.connector.connect(**db_config)

#cursor()

cursor() Method:
The cursor() method creates a cursor object, which is used to interact with the database. 
A cursor acts as a pointer that allows you to navigate through the result set of a query. 
We use the cursor to execute SQL statements and fetch data from the database.

#execute()

execute() Method:
The execute() method is used to execute SQL queries or statements using the cursor. 
We provide the SQL statement as an argument to the execute() method, and it sends the query to the database for execution. 
This method is used for various types of queries, such as SELECT, INSERT, UPDATE, DELETE, and more.
```

```python
#7.

In a typical SQL query, the clauses are executed in the following order:

SELECT: This is used to specify the columns you want to retrieve from the database.

FROM: This specifies the table or tables from which you want to retrieve the data.

JOIN: If you're using any JOIN operations to combine data from multiple tables, this is executed next.

WHERE: This filters the data based on specified conditions. Rows that meet the conditions are included in the result set.

GROUP BY: If you're using grouping, this organizes the data into groups based on specified columns.

HAVING: This filters the grouped data based on specified conditions. Similar to WHERE, but applies to grouped data.

ORDER BY: This sorts the result set based on specified columns.

LIMIT/OFFSET: If applicable, these are used to limit the number of rows returned or skip a certain number of rows.

SELECT DISTINCT: If specified, it ensures that only distinct values are returned.
```
